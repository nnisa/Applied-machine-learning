```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
dataSetUrl = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
dataSet = pd.read_csv(dataSetUrl, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
```

```python
sepal_length = dataSet['sepal_length']
sepal_width = dataSet['sepal_width']
petal_length = dataSet['petal_length']
petal_width = dataSet['petal_width']
```

```python
flower_class = dataSet['class']
colors = []
for i in flower_class:
    if i == 'Iris-setosa':
        colors.append('r');
    elif i == 'Iris-versicolor':
        colors.append('g');
    else:
        colors.append('b');
```

```python
plt.scatter(sepal_length, sepal_width, c=colors)
plt.title('sepal_length & sepal_width')
plt.savefig("sepal_length&sepal_width.png")

plt.figure()
plt.scatter(sepal_length, petal_length, c=colors)
plt.title('sepal_length & petal_length')
plt.savefig("sepal_length&petal_length.png")

plt.figure()
plt.scatter(sepal_length, petal_width, c=colors)
plt.title('sepal_length & petal_width')
plt.savefig("sepal_length&petal_width.png")

plt.figure()
plt.scatter(sepal_width, sepal_length, c=colors)
plt.title('sepal_width & sepal_length')
plt.savefig("sepal_width&sepal_length.png")

plt.figure()
plt.scatter(sepal_width, petal_length, c=colors)
plt.title('sepal_width & petal_length')
plt.savefig("sepal_width&petal_length.png")

plt.figure()
plt.scatter(sepal_width, petal_width, c=colors)
plt.title('sepal_width & petal_width')
plt.savefig("sepal_width&petal_width.png")

plt.figure()
plt.scatter(petal_length, sepal_length, c=colors)
plt.title('petal_length & sepal_length')
plt.savefig("petal_length&sepal_length.png")

plt.figure()
plt.scatter(petal_length, sepal_width, c=colors)
plt.title('petal_length & sepal_width')
plt.savefig("petal_length&sepal_width.png")

plt.figure()
plt.scatter(petal_length, petal_width, c=colors)
plt.title('petal_length & petal_width')
plt.savefig("petal_length&petal_width.png")

plt.figure()
plt.scatter(petal_width, sepal_length, c=colors)
plt.title('petal_width & sepal_length')
plt.savefig("petal_width&sepal_length.png")

plt.figure()
plt.scatter(petal_width, sepal_width, c=colors)
plt.title('petal_width & sepal_width')
plt.savefig("petal_width&sepal_width.png")

plt.figure()
plt.scatter(petal_width, petal_length, c=colors)
plt.title('petal_width & petal_length')
plt.savefig("petal_width&petal_length.png")
```
