```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd

dataset = pd.read_csv('train.csv')
dataset.head(10)
```

```python
dataset=dataset.drop(["PassengerId", "Name", "Ticket", "Cabin", "Embarked"], axis=1)
dataset.head(10)
```

```python
dataset.fillna(value=0, inplace=True)
```

```python
dataset
```

```python
dataset['Sex'] = dataset['Sex'].map({'female': 1, 'male': 0})
dataset
```

```python
X = dataset[["Pclass","Sex","Age", "Parch", "SibSp", "Fare"]]
Y = dataset[["Survived"]]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4)
```

```python
Reg = LogisticRegression()
Reg.fit(X_train, Y_train)
```

```python
Reg.fit(X_train,Y_train).score(X_test,Y_test)
```

```python
#Loading testing dataset and preparing it for prediction
TestDataset = pd.read_csv('test.csv')
resultIndex = TestDataset[["PassengerId"]]
TestDataset = TestDataset.drop(["PassengerId", "Name", "Ticket", "Cabin", "Embarked"], axis=1)
TestDataset.fillna(value=0, inplace=True)
TestDataset['Sex'] = TestDataset['Sex'].map({'female': 1, 'male': 0})
TestX = TestDataset[["Pclass","Sex","Age", "Parch", "SibSp", "Fare"]]
TestDataset
```

```python
predictions = Reg.predict(TestX)
```
